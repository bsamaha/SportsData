```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class HockeyStats:
    
    def __init__(self, name):
        self.name = name
        
    def PlotGoals(self):
        df = pd.read_csv('%s.csv' % self.name)
        x = df['GP']
        y = df['G']
        plt.scatter(x,y)
```

```python
p1 = HockeyStats("Ovechkin")
p1.PlotGoals()
```
