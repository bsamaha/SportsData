```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('modern_RAPTOR_by_player.csv')
```

```python
is_2018 =  df['season']== 2018
is_2019 =  df['season']== 2019
```

First we will parse all the players that played in 2019

```python
currentStats = df[is_2019]
```

Next will only have players who have played over 40 min

```python
topMinutes = currentStats[currentStats['mp'] > 40]
```

```python
topMinutes #488 players
```

```python
top100 = topMinutes.sort_values(by='raptor_total', ascending=False).head(100)
```

```python
top100
```

```python
bestOffense = top100[top100['raptor_box_offense'] > 0]
```
