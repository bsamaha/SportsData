```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('modern_RAPTOR_by_player.csv')
```

```python
is_2018 =  df['season']== 2018
is_2019 =  df['season']== 2019
```

```python
currentStats = df[is_2019]
```

```python
top50 = currentStats.sort_values(by='mp', ascending=False).head(50)
```
